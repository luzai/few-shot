```python
def yield_weight(t,tt):
    for name,name2,w1,w2 in zip(t.keys(),tt.keys(),t.values(),tt.values()):
        w1=w1.squeeze() 
        w2=w2.squeeze() 
        if w1.shape != w2.shape:
            assert name==name2
            yield name,w1,w2
```

```python
ttt=tt.copy()
for name,w1,w2 in yield_weight():
    mul=[]
    for m1,m2 in zip(w1.shape,w2.shape):
        mul.append(m2//m1)
    ttt[name]=np.tile(w1,mul)
```
